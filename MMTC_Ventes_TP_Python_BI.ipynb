```python
import pandas as pd
import os
# Chemin du dossier contenant les  fichiers sources
path_SO = '/content/TPBI/SO'
extensions = ['.xlsx', '.csv', '.txt', '.unl']
staging_area = {}
```

```python
# Parcourir les fichiers dans le dossier SO
for file in os.listdir(path_SO):
    if any(file.endswith(ext) for ext in extensions):
        chemin_fichier = os.path.join(path_SO, file)
        # Lire les fichiers avec Pandas
        if file.endswith('.xlsx'):
            if 'ETL-SI-SIEGE.xlsx' in file:
                excel_file = pd.ExcelFile(chemin_fichier)
                sheets = excel_file.sheet_names
                for sheet_name in sheets:
                    staging_area[f"{file}_{sheet_name}"] = pd.read_excel(excel_file, sheet_name=sheet_name)
            else:
                staging_area[file] = pd.read_excel(chemin_fichier)
        elif file.endswith('.csv'):
            staging_area[file] = pd.read_csv(chemin_fichier, encoding='latin-1', delimiter=';')
        elif file.endswith('.txt'):
            staging_area[file] = pd.read_csv(chemin_fichier, delimiter=';', encoding='latin-1')
        elif file.endswith('.unl'):
            staging_area[file] = pd.read_csv(chemin_fichier, delimiter='|', encoding='latin-1')
```

```python
for nom_fichier, dataframe in staging_area.items():
    # Afficher le nom du fichier
    print("Nom du fichier:", nom_fichier)
    # Afficher le contenu du DataFrame en utilisant display
    display(dataframe)
```

```python
staging_area_path = '/content/TPBI/SA'
# Parcourir les tables dans staging_area et les sauvegarder en tant que fichiers Excel
for table_name, table_content in staging_area.items():
    # Construire le chemin du fichier pour chaque table dans le rép de Staging Area
    file_path = os.path.join(staging_area_path, f"{table_name}.xlsx")

    # Vérifier si le contenu est un dataframe avant de sauvegarder
    if isinstance(table_content, pd.DataFrame):
        table_content.to_excel(file_path, index=False)
        print(f"Table '{table_name}' saved in Staging Area.")
    else:
        print(f"Table '{table_name}' is not a Pandas DataFrame and cannot be saved.")
```

```python
# Liste des noms de fichiers dans la zone de staging
file_names = os.listdir(staging_area_path)
# Dictionnaire pour stocker les DataFrames des tables
tables = {}
# Lire chaque fichier Excel dans la staging area
for file_name in file_names:
    # Vérifier si le fichier est au format Excel
    if file_name.endswith('.xlsx'):
        # Construire le chemin complet du fichier
        file_path = os.path.join(staging_area_path, file_name)
        # Charger le fichier Excel dans un DataFrame
        table_name = os.path.splitext(file_name)[0]
        tables[table_name] = pd.read_excel(file_path)
```

```python
# Construire la table Magasin
dim_magasin = tables['ETL-SI-SIEGE.xlsx_SI-SIEGE-Magasin'].copy()
# Fusionner avec les informations de Responsable depuis ETL-SIV-MO.unl
dim_magasin = pd.merge(dim_magasin, tables['ETL-SIV-MO.unl'][['Magasin', 'Responsable']], on='Magasin', how='left')
# Fusionner avec les informations de Seuil depuis ETL-SI-SIEGE.xlsx_SI-SIEGE-Seuil
dim_magasin = pd.merge(dim_magasin, tables['ETL-SI-SIEGE.xlsx_SI-SIEGE-Seuil'], on='Magasin', how='left')
# Ajouter la colonne 'Prix local' à dim_magasin depuis ETL-SI-SIEGE.xlsx_SI-SIEGE-Prix
dim_magasin = pd.merge(dim_magasin, tables['ETL-SI-SIEGE.xlsx_SI-SIEGE-Prix'][['Magasin', 'Prix local']],
                       on='Magasin', how='left')
# Supprimer les doublons basés sur la colonne 'Magasin'
dim_magasin = dim_magasin.drop_duplicates(subset='Magasin')
# Supprimer les indices par défaut et réinitialiser l'index
dim_magasin.reset_index(drop=True, inplace=True)
# Ajouter une colonne ID_Magasin pour les indices
dim_magasin['ID_Magasin'] = dim_magasin.index
# Déplacer la colonne 'ID_Magasin' en tant que première colonne
dim_magasin = dim_magasin[['ID_Magasin'] + [col for col in dim_magasin.columns if col != 'ID_Magasin']]
```

```python
print("Dimension Magasin:")
display(dim_magasin)
```

```python
# Construire la table Pays
dim_pays = tables['ETL-SI-SIEGE.xlsx_SI-SIEGE-Magasin'].copy()
dim_pays = dim_pays[['Pays']].drop_duplicates()
dim_pays.reset_index(drop=True, inplace=True)
# Ajouter une colonne ID_Pays
dim_pays['ID_Pays'] = dim_pays.index
dim_pays = dim_pays[['ID_Pays', 'Pays']]
print("Dimension Pays:")
display(dim_pays)
```

```python
# Construire la table Ville
dim_ville = tables['ETL-SI-SIEGE.xlsx_SI-SIEGE-Magasin'].copy()
dim_ville = dim_ville[['Ville']].drop_duplicates()
dim_ville.reset_index(drop=True, inplace=True)
# Ajouter une colonne ID_Ville
dim_ville['ID_Ville'] = dim_ville.index
dim_ville = dim_ville[['ID_Ville', 'Ville']]
print("Dimension Ville:")
display(dim_ville)
```

```python
# Construire la table Date
# Extraire les dates uniques de la table de staging
unique_dates = tables['ETL-SIV-MA.csv']['Date'].unique()
# Créer une nouvelle DataFrame pour la table "Date" en spécifiant le format de date
dim_date = pd.DataFrame({'Date': pd.to_datetime(unique_dates, format='%d/%m/%Y')})
# Ajouter la colonne "N Semaine"
dim_date['N Semaine'] = dim_date['Date'].apply(lambda x: 'S1' if x.day <= 7 else ('S2' if x.day <= 14 else ('S3' if x.day <= 21 else ('S4' if x.day <= 28 else 'S5'))))
# Charger la table ETL-SI-Banque-Taux.xlsx dans un DataFrame
taux_df = tables['ETL-SI-Banque-Taux.xlsx']
# Pivoter la table pour avoir une seule ligne par date avec Taux CAD et Taux CFA
taux_pivot = taux_df.pivot(index='Date', columns='Devise', values='Taux jour').reset_index()
# Fusionner avec la table dim_date sur la colonne 'Date'
dim_date = pd.merge(dim_date, taux_pivot, on='Date', how='left')
# Renommer les colonnes Taux CAD et Taux CFA
dim_date = dim_date.rename(columns={'CAD': 'Taux CAD', 'CFA': 'Taux CFA'})
# Ajouter une colonne ID_Date pour les indices
dim_date['ID_Date'] = dim_date.index
# Déplacer la colonne 'ID_Date' en tant que première colonne
dim_date = dim_date[['ID_Date'] + [col for col in dim_date.columns if col != 'ID_Date']]
```

```python
# La table de ventes RA
ventes_ra_df = pd.merge(tables['ETL-SIV-RA.txt'], dim_magasin[['Magasin', 'ID_Magasin']], on='Magasin', how='left')
ventes_ra_df['Date'] = pd.to_datetime(ventes_ra_df['Date'], format='%d/%m/%Y')
ventes_ra_df = pd.merge(ventes_ra_df, dim_date[['Date', 'ID_Date', 'N Semaine']], on='Date', how='left')
ventes_ra_df['ID_Ville'] = dim_ville[dim_ville['Ville'] == 'RABAT']['ID_Ville'].values[0]
ventes_ra_df['ID_Pays'] = dim_pays[dim_pays['Pays'] == 'MAROC']['ID_Pays'].values[0]
columns_order = ['ID_Magasin', 'ID_Date', 'N Semaine', 'ID_Ville', 'ID_Pays', 'Vente OK', 'Prix unité DH', 'Total DH jour', 'Retour']
ventes_ra_df = ventes_ra_df[columns_order]
ventes_ra_df.head()
```

```python
# La table de ventes CA
ventes_ca_df = pd.merge(tables['ETL-SIV-CA-Vente.csv'], dim_magasin[['Magasin', 'ID_Magasin']], on='Magasin', how='left')
ventes_ca_df['Date'] = pd.to_datetime(ventes_ca_df['Date'], format='%d/%m/%Y')
ventes_ca_df = pd.merge(ventes_ca_df, dim_date[['Date', 'ID_Date', 'N Semaine']], on='Date', how='left')
ventes_ca_df['ID_Ville'] = dim_ville[dim_ville['Ville'] == 'CASABLANCA']['ID_Ville'].values[0]
ventes_ca_df['ID_Pays'] = dim_pays[dim_pays['Pays'] == 'MAROC']['ID_Pays'].values[0]
retour_df = tables['ETL-SIV-CA-Retour.csv']
retour_df['Date'] = pd.to_datetime(retour_df['Date'], format='%d/%m/%Y')
ventes_ca_df = pd.merge(ventes_ca_df, retour_df[['Magasin', 'Date', 'Retour']], on=['Magasin', 'Date'], how='left')
ventes_ca_df['Vente OK'] = ventes_ca_df['Nbre total vente'] - ventes_ca_df['Retour']
columns_order = ['ID_Magasin', 'ID_Date', 'N Semaine', 'ID_Ville', 'ID_Pays', 'Nbre total vente', 'Observation', 'Retour', 'Vente OK']
ventes_ca_df = ventes_ca_df[columns_order]
ventes_ca_df.head()
```

```python
# La table de ventes MA
ventes_ma_df = tables['ETL-SIV-MA.csv']
ventes_ma_df['Magasin'] = 'MARR01'
ventes_ma_df['ID_Ville'] = dim_ville[dim_ville['Ville'] == 'MARRAKECH']['ID_Ville'].values[0]
ventes_ma_df['ID_Pays'] = dim_pays[dim_pays['Pays'] == 'MAROC']['ID_Pays'].values[0]
ventes_ma_df['Date'] = pd.to_datetime(ventes_ma_df['Date'], format='%d/%m/%Y')
ventes_ma_df = pd.merge(ventes_ma_df, dim_magasin[['Magasin', 'ID_Magasin']], left_on='Magasin', right_on='Magasin', how='left')
ventes_ma_df = pd.merge(ventes_ma_df, dim_date[['Date', 'ID_Date', 'N Semaine']], on='Date', how='left')
columns_order = ['ID_Magasin', 'ID_Ville', 'ID_Pays', 'ID_Date', 'N Semaine', 'Vente OK', 'Retour', 'Prix unité DH', 'Total DH jour']
ventes_ma_df = ventes_ma_df[columns_order]
ventes_ma_df.head()
```

```python
#La table de ventes MO
ventes_mo_df = tables['ETL-SIV-MO.unl']
```

```python
def convertir_format(date_str):
    try:
        # Extraire le jour
        day_str = date_str.split()[1]

        day = int(day_str)
        date_formatee = f'2023-09-{day:02}'

        return date_formatee
    except (ValueError, IndexError):
        return None

ventes_mo_df['Date'] = ventes_mo_df['Date'].apply(convertir_format)
```

```python
# Convertir la colonne 'Date' de dim_date au format datetime
ventes_mo_df['Date'] = pd.to_datetime(ventes_mo_df['Date'], errors='coerce', format='%Y-%m-%d')
# Fusionner avec les dimensions
ventes_mo_df = pd.merge(ventes_mo_df, dim_magasin[['Magasin', 'ID_Magasin']], on='Magasin', how='left')
ventes_mo_df = pd.merge(ventes_mo_df, dim_date[['Date', 'ID_Date', 'N Semaine']], on='Date', how='left')
ventes_mo_df = pd.merge(ventes_mo_df, dim_ville[dim_ville['Ville'] == 'MONTREAL'][['Ville', 'ID_Ville']], on='Ville', how='left')
ventes_mo_df = pd.merge(ventes_mo_df, dim_pays[dim_pays['Pays'] == 'CANADA'][['Pays', 'ID_Pays']], on='Pays', how='left')
columns_order = ['ID_Magasin', 'ID_Date', 'N Semaine', 'ID_Ville', 'ID_Pays', 'Total vente', 'Vente OK', 'Retour', 'Prix CAD', 'Total vente CAD', 'Observation']
ventes_mo_df = ventes_mo_df[columns_order]
ventes_mo_df.head()
```

```python
ventes_ta_subset = ventes_ta_df[['ID_Magasin', 'ID_Date', 'ID_Ville', 'ID_Pays', 'N Semaine', 'Vente OK', 'Retour']]
ventes_ra_subset = ventes_ra_df[['ID_Magasin', 'ID_Date', 'ID_Ville', 'ID_Pays', 'N Semaine', 'Vente OK', 'Retour']]
ventes_ca_subset = ventes_ca_df[['ID_Magasin', 'ID_Date', 'ID_Ville', 'ID_Pays', 'N Semaine', 'Vente OK', 'Retour']]
ventes_ma_subset = ventes_ma_df[['ID_Magasin', 'ID_Date', 'ID_Ville', 'ID_Pays', 'N Semaine', 'Vente OK', 'Retour']]
ventes_mo_subset = ventes_mo_df[['ID_Magasin', 'ID_Date', 'ID_Ville', 'ID_Pays', 'N Semaine', 'Vente OK', 'Retour']]
ventes_ou_subset = ventes_ou_df[['ID_Magasin', 'ID_Date', 'ID_Ville', 'ID_Pays', 'N Semaine', 'Vente OK', 'Retour']]
# Concaténer les tables
table_ventes = pd.concat([ventes_ta_subset, ventes_ra_subset, ventes_ca_subset, ventes_ma_subset, ventes_mo_subset, ventes_ou_subset], ignore_index=True)
# Fusionner avec la dimension 'dim_magasin' pour obtenir 'Prix Local'
table_ventes = pd.merge(table_ventes, dim_magasin[['ID_Magasin', 'Prix local']], on='ID_Magasin', how='left')
# Ajouter la colonne 'Total Ventes' en multipliant 'Vente OK' par 'Prix local'
table_ventes['Total Ventes'] = table_ventes['Vente OK'] * table_ventes['Prix local']
table_ventes = table_ventes.drop(columns=['Prix local'])
#table_ventes = table_ventes[table_ventes['N Semaine'] != 'S5']
# Fusionner avec la dimension 'dim_date' pour obtenir les taux de conversion
table_ventes = pd.merge(table_ventes, dim_date[['ID_Date', 'Taux CAD', 'Taux CFA']], on='ID_Date', how='left')
# Appliquer le taux de conversion en fonction du magasin
table_ventes['Total Ventes (DH)'] = table_ventes.apply(
    lambda row: row['Total Ventes'] * row['Taux CAD'] if row['ID_Magasin'] in [4, 5] else
                  row['Total Ventes'] * row['Taux CFA'] if row['ID_Magasin'] in [6, 7] else
                  row['Total Ventes'],
    axis=1
)
table_ventes = table_ventes.drop(columns=['Taux CAD', 'Taux CFA', 'Total Ventes'])
```

```python
table_ventes_canada = table_ventes[table_ventes['ID_Pays'] == 1]
ventes_par_mag = pd.merge(table_ventes_canada, dim_magasin[['ID_Magasin', 'Magasin']], on='ID_Magasin', how='left')
ventes_par_mag_total = ventes_par_mag.groupby('Magasin')['Vente OK'].sum().reset_index()
ventes_par_mag_total['Pourcentage'] = ((ventes_par_mag_total['Vente OK'] / ventes_par_mag_total['Vente OK'].sum()) * 100).round(0)
colors = ['#0566FF', '#50E3C2']
fig_pie2 = px.pie(ventes_par_mag_total, names='Magasin', values='Pourcentage',
                 title='Part des magasins en unité',
                 labels={'Pourcentage': 'Pourcentage de Ventes'},
                 hover_data=['Pourcentage'],
                 color_discrete_sequence=colors)
fig_pie2.update_traces(textinfo='label+percent')
fig_pie2.update_layout(title_x=0.5)
fig_pie2.update_layout(width=500, height=350)
fig_pie2.show()
```

```python
table_ventes_bf = table_ventes[table_ventes['ID_Pays'] == 2]
ventes_par_mag2 = pd.merge(table_ventes_bf, dim_magasin[['ID_Magasin', 'Magasin']], on='ID_Magasin', how='left')
ventes_par_mag_total2 = ventes_par_mag2.groupby('Magasin')['Vente OK'].sum().reset_index()
ventes_par_mag_total2['Pourcentage'] = ((ventes_par_mag_total2['Vente OK'] / ventes_par_mag_total2['Vente OK'].sum()) * 100).round(0)
colors = ['#0566FF', '#50E3C2']
fig_pie3 = px.pie(ventes_par_mag_total2, names='Magasin', values='Pourcentage',
                 title='Part des magasins en unité',
                 labels={'Pourcentage': 'Pourcentage de Ventes'},
                 hover_data=['Pourcentage'],
                 color_discrete_sequence=colors)
fig_pie3.update_traces(textinfo='label+percent')
fig_pie3.update_layout(title_x=0.5)
fig_pie3.update_layout(width=500, height=350)
fig_pie3.show()
```

```python
ventes_par_ville_total = table_ventes.groupby('ID_Ville')['Total Ventes (DH)'].sum().reset_index()
ventes_par_ville_total = pd.merge(ventes_par_ville_total, dim_ville[['ID_Ville', 'Ville']], on='ID_Ville', how='left')
ventes_par_ville_total = ventes_par_ville_total.sort_values(by='Total Ventes (DH)', ascending=False)
colors = px.colors.sequential.Turbo
fig_bar = px.bar(ventes_par_ville_total, x='Ville', y='Total Ventes (DH)', color='Ville',
                 title='Montant kdh', color_discrete_sequence=colors)
fig_bar.update_xaxes(linecolor="#D8D9DA")
fig_bar.update_layout(title_x=0.5, xaxis_title=None, yaxis_title=None, showlegend=False)
fig_bar.update_layout(width=500, height=400)
fig_bar.show()
```

```python
ventes_par_pays_total = table_ventes.groupby('ID_Pays')['Total Ventes (DH)'].sum().reset_index()
ventes_par_pays_total = pd.merge(ventes_par_pays_total, dim_pays[['ID_Pays', 'Pays']], on='ID_Pays', how='left')
ventes_par_pays_total['Total Ventes (KDH)'] = ventes_par_pays_total['Total Ventes (DH)'] / 1000
ventes_par_pays_total = ventes_par_pays_total.sort_values(by='Total Ventes (KDH)', ascending=False)
colors = px.colors.sequential.Turbo
fig_bar_pays = px.bar(ventes_par_pays_total, x='Pays', y='Total Ventes (KDH)',
                      title='Montant kdh',
                      hover_data={'Total Ventes (KDH)': ':,.0f'},
                      color='Pays',
    color_discrete_map={country: color for country, color in zip(ventes_par_pays_total['Pays'], colors)})
fig_bar_pays.update_xaxes(linecolor="#D8D9DA")
fig_bar_pays.update_layout(title_x=0.5, xaxis_title=None, yaxis_title=None, showlegend=False)
fig_bar_pays.update_layout(width=500, height=400)
fig_bar_pays.show()
```

```python
taux_retour_par_magasin = table_ventes.groupby('ID_Magasin').agg({'Retour': 'sum', 'Vente OK': 'sum'}).reset_index()
taux_retour_par_magasin['Taux de Retour'] = (taux_retour_par_magasin['Retour'] / taux_retour_par_magasin['Vente OK'] * 100)
taux_retour_par_magasin = pd.merge(taux_retour_par_magasin, dim_magasin[['ID_Magasin', 'Magasin']], on='ID_Magasin', how='left')
taux_retour_par_magasin =taux_retour_par_magasin.sort_values(by='Taux de Retour', ascending=False)
colors = ['#07BEB8']
fig_bar_taux_retour = px.bar(taux_retour_par_magasin, x='Taux de Retour', y='Magasin',
                             title='Taux retour (%)',
                             color='Magasin',
                             labels={'Taux de Retour': 'Taux de Retour (%)'},
                             orientation='h',
                             hover_data={'Retour': ':,.2f'},
                             color_discrete_sequence=colors)
fig_bar_taux_retour.update_layout(title_x=0.5, xaxis_title=None, yaxis_title=None, showlegend=False)
fig_bar_taux_retour.update_xaxes(linecolor="#D8D9DA")
fig_bar_taux_retour.update_yaxes(linecolor="#D8D9DA")
fig_bar_taux_retour.update_layout(width=500, height=400)
fig_bar_taux_retour.show()
```
